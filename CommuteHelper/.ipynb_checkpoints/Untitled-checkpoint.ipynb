```python
import os
import csv
import pandas as pd
from sqlalchemy import create_engine
```

```python
gtfs_dir = r'C:\Users\Katie\Documents\courses\CommuteHelper\MBTA_GTFS'
stop_filename = os.path.join(gtfs_dir,'stops.txt')
trips_filename = os.path.join(gtfs_dir,'trips.txt')
shapes_filename = os.path.join(gtfs_dir,'shapes.txt')
routes_filename = os.path.join(gtfs_dir,'routes.txt')
```

```python
stops = pd.read_csv(stop_filename)
trips = pd.read_csv(trips_filename)
shapes = pd.read_csv(shapes_filename)
routes = pd.read_csv(routes_filename)
```

```python

engine = create_engine('postgresql://postgres:postgres@localhost:5432/mbta_data')
```

```python
stops.to_sql('stops', engine, if_exists='append')
```

```python
stops.to_sql('stops', engine)
trips.to_sql('trips', engine)
shapes.to_sql('shapes', engine)
routes.to_sql('routes', engine)
```
